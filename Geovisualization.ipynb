CREATING A SIMPLE MAP

```python

# Import necessary libraries
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the GeoJSON data
file_path = '/content/drive/MyDrive/Data/Mombasa.geojson'
mombasa = gpd.read_file(file_path)

# Plot the data
mombasa.plot(figsize=(10, 6), color='lightblue', edgecolor='black')
plt.title("Mombasa Map")
plt.show()
```

Adding More elements

```python
# Import necessary libraries
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx  # for basemap
# Load the GeoJSON data
file_path = '/content/drive/MyDrive/Data/Mombasa.geojson'
mombasa = gpd.read_file(file_path)
# Create a plot with the data
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
mombasa.plot(ax=ax, color='lightblue', edgecolor='black')

# Add a basemap using contextily
ctx.add_basemap(ax, crs=mombasa.crs.to_string(), source=ctx.providers.CartoDB.Positron)

# Add title and labels
plt.title("Mombasa Map", fontsize=16)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Add gridlines for reference
ax.grid(True, linestyle='--', alpha=0.7)

# Show the map
plt.show()
```
